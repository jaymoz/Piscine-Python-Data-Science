## importing necessary tools

```python
import pandas as pd 
import sqlite3
```

## create a connection to the database using the library sqlite3

```python
db = sqlite3.connect("../data/checking-logs.sqlite.sqlite_copy")
```

## create a new table datamart in the database by joining the tables pageviews and checker using only one query

```python
data = pd.io.sql.read_sql('select * from checker limit 10', db)
data
```

```python
query = '''
select checker.uid as uid, labname, min(checker.timestamp) as first_commit_ts, min(pageviews.datetime) as first_view_ts
from checker left join pageviews on checker.uid = pageviews.uid
where
    labname in ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1') and
    checker.uid not like "admin%" and
    status = 'ready' and
    numTrials = 1
group by checker.uid, labname

'''
datamart = pd.io.sql.read_sql(query, db, parse_dates=['first_commit_ts', 'first_view_ts'])
datamart
```

## using Pandas methods, create two dataframes: test and control

```python
test = datamart.dropna()
test
```

```python
control = datamart[datamart['first_view_ts'].isna()]
control
```

```python
control = control.fillna(test['first_view_ts'].mean())
control
```

```python
pd.io.sql.to_sql(test, 'test', db, if_exists='replace')
```

```python
pd.io.sql.to_sql(control, 'control', db, if_exists='replace')
```

## close the connection

```python
db.close()
```
