## Imports

```python
import pandas as pd 
import sqlite3 
```

```python
db = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
```

## analyze only the users and not the admins 

```python
query = '''
SELECT uid, datetime FROM pageviews
WHERE uid LIKE 'user_%'
'''

views = pd.io.sql.read_sql(query, db, parse_dates=['datetime'])
views
```

```python
query = '''
SELECT uid, timestamp FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, db, parse_dates=['timestamp'])
commits
```

## analyze only the dates when there were both views and checker commits

```python
group_views = views.groupby(by=views['datetime'].dt.date).count()['uid'].reset_index()
group_views.columns = ['date', 'views']
group_views = group_views.set_index('date')
group_views
```

```python
group_commits = commits.groupby(by=commits['timestamp'].dt.date).count()['uid'].reset_index()
group_commits.columns = ['date', 'commits']
group_commits = group_commits.set_index('date')
group_commits
```

```python
views_commits = group_views.merge(group_commits,how='inner',left_on=group_views.index,right_on=group_commits.index).set_index('key_0')
views_commits.index = views_commits.index.rename('date')
views_commits
```

## use size of the font should be 8, size of the figure is (15, 8)

```python
fig = views_commits.plot(fontsize=8, figsize=(15,8),title='Views and commits per day', color=['red', 'c'], style=['--', '-.'], rot=90)
fig.grid(axis='x')
fig.get_lines()[0].set_marker('o')
fig.get_legend().get_lines()[0].set_marker('o')
fig.set_facecolor('#F5F5F5')
```
